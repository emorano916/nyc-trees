```python
import pandas as pd
import numpy as np 
```

```python
#opening the csv
dataset = pd.read_csv('data_100000.csv')
dataset.head()
```

```python
#Getting the sum of the NaN values 
dataset.isnull().sum()
```

```python
#Replace the NaN value by 'Unknow' in order to keep the rest of rows datas.
datasetLT = dataset.fillna('Unknown')
datasetLT
```

```python
#checking of dtype
datasetLT.dtypes
```

```python
#checking if there is dataset
datasetLT.duplicated().value_counts()
```

```python
#checking of the shape
datasetLT.shape
```

```python
#checking of statistical datas
datasetLT.describe(include='all').T
```

```python
#Checking of the value of stump_diam
datasetLT['stump_diam'].value_counts()
```

```python
#check if columns might have problems and  are consolidated
datasetLT.zip_city.unique()
datasetLT.boroname.unique()
datasetLT.user_type.unique()
datasetLT.guards.unique()
datasetLT.steward.unique()
datasetLT.created_at.unique()
datasetLT.health.unique()
datasetLT.nta_name.unique()
```

```python
#drop of the state columns (only one New-York value) and the_geom because we have lat and lon in 2 others columns
datasetLT = datasetLT.drop(['state', 'the_geom'], axis=1)
datasetLT
```

```python
#Transform of the datetime object to datetime type
datasetLT['created_at'] = pd.to_datetime(datasetLT['created_at'], 
 format = '%m/%d/%Y')

#Creation of 3 new columns
datasetLT['Year']=datasetLT['created_at'].dt.year
datasetLT['Month']=datasetLT['created_at'].dt.month
datasetLT['Day']=datasetLT['created_at'].dt.day
datasetLT
```

```python
#get rid of the left and right space in all the dataset
datasetLT.columns = datasetLT.columns.str.strip()
```

```python
datasetLT.dtypes
```

```python
#tranforming of one column with OrdinalEncoder of Scikit Learn 
#that ML algorithms will assume that two nearby
#values are more similar than two distant values. This may be fine in some cases (e.g.,
#for ordered categories such as “bad”, “average”, “good”, “excellent”)
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
datasetLT_health_cat = datasetLT[['health']]
datasetLT_health_cat_encoded = ordinal_encoder.fit_transform(datasetLT_health_cat)
datasetLT_health_cat_encoded[:10]
   
#Check of categories encoded
ordinal_encoder.categories_
```

```python
datasetLT[['health']] = datasetLT_health_cat_encoded
datasetLT.head()
```

```python
#checking of the value of curb_loc
datasetLT.curb_loc.unique()
```

```python
#Apply yes to all OnCurb values with a lambda function
datasetLT['curb_loc'] = datasetLT['curb_loc'].apply(lambda x : "yes" if 'OnCurb' in x else "no")
datasetLT['curb_loc']
```

```python
#Apply get_dummies to replace Yes, No by binary values 
df = pd.get_dummies(datasetLT.curb_loc, prefix='OnCurb', drop_first="True")
df.rename(columns={"OnCurb_yes" : "OnCurb"})
datasetLT['curb_loc'] = df
```

```python
datasetLT.head()
```

```python
#Getting binary values of nta_name columns with OneHotEncoder of Scikit Learn
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
datasetLT_nta_name_cat = datasetLT[['nta_name']]
datasetLT_nta_name_cat_1hot = cat_encoder.fit_transform(datasetLT_nta_name_cat)

#check of the matrix type (spare matrix)
datasetLT_nta_name_cat_1hot
```

```python
#Check of categories encoded
cat_encoder.categories_
```

```python
#Join the new columns to the dataset
datasetLT= datasetLT.join(datasetLT_nta_name_cat_1hot)
datasetLT.head()
```

```python
#New test with OrdinalEncoder
datasetLT_nta_name_cat_encoded = ordinal_encoder.fit_transform(datasetLT_nta_name_cat)
```

```python
datasetLT[['nta_name']] = datasetLT_nta_name_cat_encoded
datasetLT.head()
```
